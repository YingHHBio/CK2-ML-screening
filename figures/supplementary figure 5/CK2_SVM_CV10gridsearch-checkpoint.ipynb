```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
from scipy.stats import spearmanr, pearsonr
from lifelines.utils import concordance_index
from mapie.regression import MapieRegressor
from joblib import parallel_backend
from pathlib import Path
from sklearn.model_selection import train_test_split
import pickle
parameter_path= Path("/Users/huiyan/Library/CloudStorage/OneDrive-UniversityofHelsinki/PhD project/CK2ab-3rd/Intergrative active/redeal/7th/bioactivity/parameter")
plot_path= Path("/Users/huiyan/Library/CloudStorage/OneDrive-UniversityofHelsinki/PhD project/CK2ab-3rd/Intergrative active/redeal/7th/bioactivity/plot")
data_path = Path("/Users/huiyan/Library/CloudStorage/OneDrive-UniversityofHelsinki/PhD project/CK2ab-3rd/Intergrative active/redeal/7th/bioactivity")
```

```python
df_dataset = pd.read_csv(data_path/"Kd_formodel_fp_standard.csv",sep = ",")
df_dataset
```

```python
df_dataset.rename(columns={"is_cell.based": "is_cellbased"}, inplace=True)
df_dataset
```

```python
X = df_dataset.drop(["Unnamed: 0","realsmiles", "is_biochemical", "Log10", "is_cellbased", "id", "type", "value"],axis=1)
X
```

```python
y = df_dataset.Log10
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.75, random_state = 0)
```

```python
X_train=X_train.values
y_train=y_train.values
X_test=X_test.values
y_test=y_test.values
```

```python
X_train
```

```python
# Define SVR model, default
svr_reg = SVR(kernel='rbf', C=1.0, gamma='scale')
```

```python
# Fit SVR model on training data
svr_reg.fit(X_train, y_train)
```

```python
y_pred_train= svr_reg.predict(X_train)
```

```python
mse = mean_squared_error(y_train, y_pred_train)
rmse = np.sqrt(mse)
nrmse = rmse / (np.max(y_train) - np.min(y_train))
```

```python
print("Spearman correlation:", spearmanr(y_train, y_pred_train))
print("Pearson correlation:", pearsonr(y_train, y_pred_train))
print("RMSE:", rmse)
print("CI:", concordance_index(y_train, y_pred_train))
print("NRMSE:", nrmse)
```

```python
y_pred_test= svr_reg.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred_test)
rmse = np.sqrt(mse)
nrmse = rmse / (np.max(y_test) - np.min(y_test))
```

```python
print("Spearman correlation:", spearmanr(y_test, y_pred_test))
print("Pearson correlation:", pearsonr(y_test, y_pred_test))
print("RMSE:", rmse)
print("CI:", concordance_index(y_test, y_pred_test))
print("NRMSE:", nrmse)
```

# 模型评估

```python
# 定义要进行调整的参数范围
```

```python
param_grid = {
    'C': [0.1, 1, 10], 
    'kernel':('linear', 'rbf'),
    'gamma': [1, 0.1, 0.01]
} 
```

```python
# 使用GridSearchCV函数进行参数调整
```

```python
grid_search = GridSearchCV(
    scoring="neg_mean_squared_error",
    estimator=svr_reg,
    param_grid=param_grid,
    cv=10,  # 交叉验证折数
    n_jobs=-1  # 并行处理的CPU核心数量，-1表示使用所有可用的CPU核心
)
```

```python
# 在训练集上进行参数搜索
```

```python
grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_params_, -grid_search.best_score_)
```

```python
cv_results = pd.DataFrame(grid_search.cv_results_)
cv_results
```

```python
cv_results.to_csv(parameter_path/"SVR_Kd.csv",sep = ",")
```

```python
# 输出最佳参数组合和最佳得分
```

```python
print("Best parameters: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
```

```python
# 使用最佳参数组合评估模型在测试集上的性能
```

```python
final_svr_model = grid_search.best_estimator_
```

```python
# # 评估模型
```

```python
y_pred_train= final_svr_model.predict(X_train)
```

```python
mse = mean_squared_error(y_train, y_pred_train)
rmse = np.sqrt(mse)
nrmse = rmse / (np.max(y_train) - np.min(y_train))
```

```python
print("Spearman correlation:", spearmanr(y_train, y_pred_train))
print("Pearson correlation:", pearsonr(y_train, y_pred_train))
print("RMSE:", rmse)
print("CI:", concordance_index(y_train, y_pred_train))
print("NRMSE:", nrmse)
```

```python
y_pred_test= final_svr_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred_test)
rmse = np.sqrt(mse)
nrmse = rmse / (np.max(y_test) - np.min(y_test))
```

```python
print("Spearman correlation:", spearmanr(y_test, y_pred_test))
print("Pearson correlation:", pearsonr(y_test, y_pred_test))
print("RMSE:", rmse)
print("CI:", concordance_index(y_test, y_pred_test))
print("NRMSE:", nrmse)
```
